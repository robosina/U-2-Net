```python
import cv2
import glob
import numpy as np
import matplotlib.pyplot as plt
def read_sort(path):
    file=np.sort(glob.glob(path + '*'))
    return file
def read_col_image(paths,name):
    imgs=np.zeros((200,400,3)).astype(np.uint8)+90
    cv2.putText(imgs,name,(80,100),0,1.3,(255,255,255),2,cv2.LINE_AA)
    imgs[:,0]=255
    imgs[:,399]=255
    for k in paths:
        img=cv2.imread(k)
        img=cv2.resize(img,(400,400))
        if(len(imgs)==0):
            imgs=img
        else:
            imgs=cv2.vconcat([imgs,img])
    return imgs
original=read_sort('D:/_masoomi/unet/U-2-Net/test_data/test_images/')
iman=read_sort('D:/_masoomi/unet/U-2-Net/test_data/iman/')
p2net_small=read_sort('D:/_masoomi/unet/U-2-Net/test_data/u2netp_results/')
p2net_big=read_sort('D:/_masoomi/unet/U-2-Net/test_data/u2net_results/')
```

```python
ori=read_col_image(original,'Original')
im=read_col_image(iman,'iman')
p2nets=read_col_image(p2net_small,'small model')
p2netb=read_col_image(p2net_big,'big model')

result=cv2.hconcat([ori,im])
result=cv2.hconcat([result,p2nets])
result=cv2.hconcat([result,p2netb])
result=cv2.cvtColor(result,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,15))
plt.imshow(result)
result=cv2.cvtColor(result,cv2.COLOR_BGR2RGB)
cv2.imwrite('D:/_masoomi/unet/U-2-Net/final.jpg',result)
```
